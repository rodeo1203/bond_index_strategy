```python
import pandas as pd

# Read each Excel file individually, skipping the first 6 rows
df1 = pd.read_excel("iBoxx USD Sustainability-Linked Bonds_1Y_Total Return Index Level.xlsx", skiprows=6)
df2 = pd.read_excel("iBoxx USD Corporates Fallen Angels_1Y_Total Return Index Level.xlsx", skiprows=6)
df3 = pd.read_excel("iBoxx JUST Capital USD High Yield Benchmark_1Y_Total Return Index Level.xlsx", skiprows=6)
df4 = pd.read_excel("iBoxx $ Treasuries_1Y_Total Return Index Level.xlsx", skiprows=6)
df5 = pd.read_excel("Markit iBoxx USD FRN Investment Grade Corporates_1Y_Total Return Index Level.xlsx", skiprows=6)
df6 = pd.read_excel("iBoxx USD Liquid High Yield Energy (5_25_50 Issuer Cap)_1Y_Total Return Index Level.xlsx", skiprows=6)
df7 = pd.read_excel("iBoxx USD High Yield Developed Markets_1Y_Total Return Index Level.xlsx", skiprows=6)
df8 = pd.read_excel("iBoxx $ Overall_1Y_Total Return Index Level.xlsx", skiprows=6)

# Keep only relevant columns and rename 'Total Return Index Level' to the index name
df1 = df1[['Date', 'Total Return Index Level']].rename(columns={'Total Return Index Level': 'USD Sustainability-Linked Bonds'})
df2 = df2[['Date', 'Total Return Index Level']].rename(columns={'Total Return Index Level': 'USD Corporates Fallen Angels'})
df3 = df3[['Date', 'Total Return Index Level']].rename(columns={'Total Return Index Level': 'JUST Capital USD High Yield Benchmark'})
df4 = df4[['Date', 'Total Return Index Level']].rename(columns={'Total Return Index Level': 'Treasuries'})
df5 = df5[['Date', 'Total Return Index Level']].rename(columns={'Total Return Index Level': 'USD FRN Investment Grade Corporates'})
df6 = df6[['Date', 'Total Return Index Level']].rename(columns={'Total Return Index Level': 'USD Liquid High Yield Energy'})
df7 = df7[['Date', 'Total Return Index Level']].rename(columns={'Total Return Index Level': 'USD High Yield Developed Markets'})
df8 = df8[['Date', 'Total Return Index Level']].rename(columns={'Total Return Index Level': 'Overall'})

# Merge all dataframes on the 'Date' column
merged_df = pd.merge(df1, df2, on='Date', how='outer')
merged_df = pd.merge(merged_df, df3, on='Date', how='outer')
merged_df = pd.merge(merged_df, df4, on='Date', how='outer')
merged_df = pd.merge(merged_df, df5, on='Date', how='outer')
merged_df = pd.merge(merged_df, df6, on='Date', how='outer')
merged_df = pd.merge(merged_df, df7, on='Date', how='outer')
merged_df = pd.merge(merged_df, df8, on='Date', how='outer')

# Convert 'Date' column to datetime format for easier time series handling
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

# Sort by date
merged_df = merged_df.sort_values(by='Date').reset_index(drop=True)

# Display the merged DataFrame
merged_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure Date column is converted to datetime and set as index
merged_df['Date'] = pd.to_datetime(merged_df['Date'])
merged_df.set_index('Date', inplace=True)  # Set Date as the index for time series analysis

# Step 1: Calculate daily returns for each sector and overall index
returns_df = merged_df.pct_change().dropna()  # Calculate percentage change and remove NaN values

# Display statistical summary of returns
print("Statistical Summary of Returns (excluding Date):")
print(returns_df.describe())

# Plot correlation matrix of returns
plt.figure(figsize=(10, 8))
sns.heatmap(returns_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Credit Sector Returns (excluding Date)')
plt.show()

# Step 2: Calculate Annualized Returns and Sharpe Ratios

# Exclude 'Overall' from sectors list for pairwise calculations with it
sectors = returns_df.columns.drop('Overall')

# Define number of trading days and risk-free rate
trading_days = 252  # Assume 252 trading days in a year
risk_free_rate = 0.04  # Assume 4% annual risk-free rate

# Calculate annualized return, standard deviation, and Sharpe ratio for each sector
annualized_returns = returns_df[sectors].mean() * trading_days
annualized_std_devs = returns_df[sectors].std() * np.sqrt(trading_days)
annualized_sharpe_ratios = (annualized_returns - risk_free_rate) / annualized_std_devs

# Plot bar chart for annualized returns of each sector
plt.figure(figsize=(12, 6))
plt.bar(annualized_returns.index, annualized_returns.values, color='blue', edgecolor='black', alpha=0.7)
plt.title('Annualized Return of Each Sector')
plt.xlabel('Sector')
plt.ylabel('Annualized Return')
plt.xticks(rotation=45)
plt.show()

# Plot bar chart for annualized standard deviations of each sector
plt.figure(figsize=(12, 6))
plt.bar(annualized_std_devs.index, annualized_std_devs.values, color='orange', edgecolor='black', alpha=0.7)
plt.title('Annualized Standard Deviation of Each Sector')
plt.xlabel('Sector')
plt.ylabel('Annualized Standard Deviation')
plt.xticks(rotation=45)
plt.show()

# Plot bar chart for annualized Sharpe Ratios of each sector
plt.figure(figsize=(12, 6))
plt.bar(annualized_sharpe_ratios.index, annualized_sharpe_ratios.values, color='green', edgecolor='black', alpha=0.7)
plt.title('Annualized Sharpe Ratio of Each Sector')
plt.xlabel('Sector')
plt.ylabel('Sharpe Ratio (Risk-Free Rate = 4%)')
plt.xticks(rotation=45)
plt.show()


```

```python
for sector in sectors:
    merged_df[f'{sector}_Spread'] = merged_df[sector] - merged_df['Overall']

# Step 4: Calculate Z-Score for Mean Reversion Strategy
# Use a rolling window to calculate the rolling mean and standard deviation of each spread
rolling_window = 30  # Adjust based on your preference
for sector in sectors:
    spread_col = f'{sector}_Spread'
    merged_df[f'{spread_col}_Mean'] = merged_df[spread_col].rolling(window=rolling_window).mean()
    merged_df[f'{spread_col}_Std'] = merged_df[spread_col].rolling(window=rolling_window).std()
    merged_df[f'{spread_col}_Z'] = (merged_df[spread_col] - merged_df[f'{spread_col}_Mean']) / merged_df[f'{spread_col}_Std']

# Step 5: Generate Trade Signals Based on Z-Score
# Define a Z-score threshold to trigger trade signals
z_score_threshold = 1.5  # Customize based on backtesting results

for sector in sectors:
    spread_z = f'{sector}_Spread_Z'
    merged_df[f'{sector}_Trade_Signal'] = np.where(merged_df[spread_z] > z_score_threshold, f'Sell {sector} / Buy Overall',
                                                   np.where(merged_df[spread_z] < -z_score_threshold, f'Buy {sector} / Sell Overall', 'Hold'))

# Step 6: Plot the Z-Score and Trade Signals for Each Sector
plt.figure(figsize=(14, 8))
for sector in sectors:
    plt.plot(merged_df.index, merged_df[f'{sector}_Spread_Z'], label=f'{sector} Spread Z-Score')
    plt.axhline(z_score_threshold, color='red', linestyle='--', label='Sell Threshold' if sector == sectors[0] else "")
    plt.axhline(-z_score_threshold, color='green', linestyle='--', label='Buy Threshold' if sector == sectors[0] else "")
plt.axhline(0, color='black', linestyle='-')
plt.legend()
plt.title('Sector Price Spreads Z-Score with Trade Signals')
plt.xlabel('Date')
plt.ylabel('Z-Score')
plt.show()

# Step 7: Display Trade Signals
# This will show trade signals where a buy or sell signal was generated
for sector in sectors:
    trade_signals = merged_df[merged_df[f'{sector}_Trade_Signal'] != 'Hold']
    print(f"\nTrade Signals for {sector}:")
    print(trade_signals[['Overall', sector, f'{sector}_Spread', f'{sector}_Spread_Z', f'{sector}_Trade_Signal']])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Parameters for backtesting
initial_capital = 1000000  # Initial capital in dollars
position_size = 100000  # Position size per trade in dollars
z_score_threshold = 1.5 # Z-score threshold for generating trade signals
risk_free_rate = 0.04  # Risk-free rate for Sharpe ratio calculation

# Create a DataFrame to track the portfolio value over time
portfolio_df = pd.DataFrame(index=merged_df.index)
portfolio_df['Cash'] = initial_capital
portfolio_df['Position'] = 0  # Tracks the cumulative P&L from all closed positions
portfolio_df['Total_Value'] = portfolio_df['Cash']  # Total portfolio value

# Track individual trade profits and signals
trade_history = []

# Backtest each sector based on the trade signals
for sector in sectors:
    spread_col = f'{sector}_Spread'
    trade_signal_col = f'{sector}_Trade_Signal'
    spread_z_col = f'{sector}_Spread_Z'

    # Initialize position and profit tracking for each sector
    position = 0  # +1 for "Buy Sector / Sell Overall", -1 for "Sell Sector / Buy Overall"
    entry_price = 0  # Entry price for the current position
    entry_date = None  # Track entry date for holding period

    for date, row in merged_df.iterrows():
        # Get the current spread and trade signal
        spread = row[spread_col]
        trade_signal = row[trade_signal_col]

        # Check for trade entry signal
        if trade_signal == f'Buy {sector} / Sell Overall' and position == 0:
            position = 1  # Enter long position
            entry_price = spread
            entry_date = date  # Record entry date
            trade_history.append((date, sector, 'Buy', entry_price))
        
        elif trade_signal == f'Sell {sector} / Buy Overall' and position == 0:
            position = -1  # Enter short position
            entry_price = spread
            entry_date = date  # Record entry date
            trade_history.append((date, sector, 'Sell', entry_price))

        # Track daily PnL but do not update entry price
        if position != 0:
            # Calculate daily PnL only for tracking purposes
            daily_pnl = position * (spread - entry_price) * position_size

            # Check for exit condition
            if (position == 1 and row[spread_z_col] >= -z_score_threshold) or \
               (position == -1 and row[spread_z_col] <= z_score_threshold):
                
                # Calculate final PnL for the trade and holding period
                exit_price = spread
                holding_period = (date - entry_date).days
                pnl = position * (exit_price - entry_price) * position_size  # Calculate PnL on exit
                trade_history.append((date, sector, 'Exit', exit_price, pnl, holding_period))
                
                # Update cumulative profit in portfolio only on trade exit
                portfolio_df.loc[date, 'Position'] += pnl
                position = 0  # Close the position

# Calculate final portfolio values
portfolio_df['Total_Value'] = portfolio_df['Cash'] + portfolio_df['Position'].cumsum()

# Plot the portfolio performance over time
plt.figure(figsize=(14, 8))
plt.plot(portfolio_df['Total_Value'], label='Total Portfolio Value')
plt.title('Portfolio Value Over Time')
plt.xlabel('Date')
plt.ylabel('Portfolio Value ($)')
plt.legend()
plt.show()

# Calculate performance metrics
total_profit = portfolio_df['Total_Value'][-1] - initial_capital
annualized_return = (portfolio_df['Total_Value'][-1] / initial_capital) ** (252 / len(portfolio_df)) - 1
annualized_volatility = portfolio_df['Total_Value'].pct_change().std() * np.sqrt(252)
sharpe_ratio = (annualized_return - risk_free_rate) / annualized_volatility
max_drawdown = (portfolio_df['Total_Value'].cummax() - portfolio_df['Total_Value']).max()

print(f"Total Profit: ${total_profit:.2f}")
print(f"Annualized Return: {annualized_return * 100:.2f}%")
print(f"Annualized Volatility: {annualized_volatility * 100:.2f}%")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Maximum Drawdown: ${max_drawdown:.2f}")

# Output trade history to a DataFrame and save to CSV
trade_history_df = pd.DataFrame(trade_history, columns=['Date', 'Sector', 'Action', 'Spread', 'PnL', 'Holding_Period'])
trade_history_df['Date'] = pd.to_datetime(trade_history_df['Date'])  # Ensure date format is consistent
trade_history_df.set_index('Date', inplace=True)
trade_history_df.to_csv('trade_history_with_pnl.csv')

print("\nTrade history with PnL and holding period saved to 'trade_history_with_pnl.csv'")

```
